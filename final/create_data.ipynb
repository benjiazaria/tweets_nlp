```python
id_tweet = pd.read_csv('data/id_tweet.tsv', sep='\t', header=0)
```

```python
clean_tweets=[]
tweets = id_tweet['tweet'].values.tolist()
for tweet in tweets:
    temp_tweet = re.sub('( @\w+|@\w+)', '', tweet)

    clean_tweets.append(re.sub('[^A-Za-z0-9\',\. ]+',"",temp_tweet))


id_tweet['tweet'] = clean_tweets
```

```python
df_unpivot_comp['variable']=df_unpivot_comp['variable'].astype(int)
id_tweet['id']=id_tweet['id'].astype(int)
id_tweet
```

```python
with_tweet = pd.merge(df_unpivot_comp, id_tweet, how='left', left_on = 'variable', right_on = 'id')
```

```python
final_data = with_tweet[['Q21','Q22','Q23','Q24','Q25','Q26','Q28','Q29','Q30','Q31','tweet','value']]
train_data = final_data[['Q21','Q22','Q23','Q24','Q25','Q26','Q28','Q29','Q30','Q31','tweet','value']][:7002]
test_data = final_data[['Q21','Q22','Q23','Q24','Q25','Q26','Q28','Q29','Q30','Q31','tweet','value']][7002:]
```
